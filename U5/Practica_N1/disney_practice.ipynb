```python
# DISNEY PRACTICES

# Importar librerías
import pandas as pd

# Leer el .csv
df_disney = pd.read_csv('Copia-de-disney_plus_titles.csv')
df_disney
```

```python
# Obtener información del DataSet.
print("--DESCRIPCIÓN DEL DATASET--")
print(df_disney.describe())
print("--INFO DEL DATASET--")
print(df_disney.info())
```

```python
# 1.Mostrar cuántos valores nulos hay por columna.
print("¿Cuantos Valores Nulos hay?")
print(df_disney.isna().sum())
```

```python
# 2.Reemplazar valores nulos en rating por "No Rating".
df_disney['rating'] = df_disney['rating'].fillna('No Rating')
```

```python
# 3.Filtrar todos los títulos agregados a la plataforma en el año 2021.
titles_2021 = df_disney.loc[df_disney['release_year'] == 2021]
titles_2021
```

```python
# 4.Contar cuántos títulos hay de cada tipo (Movie, TV Show).

# El atributo Shape permite conocer la dimensión de un DataFrame. 
# Shape[0] = Cantidad de filas.
# Shape[1] = Cantidad de columnas.
count_Movie = df_disney.loc[df_disney['type'] == 'Movie'].shape[0]
print("La cantidad de títulos del tipo Movie son:", count_Movie)
```

```python
# 5.Crear una nueva columna que indique si el título fue lanzado antes o después del año 2000.
df_disney['Lanzamiento'] = df_disney['release_year'].apply(lambda x: 'After' if x > 2000 else 'Before')
df_disney
```

```python
# 6.Usar .map() para traducir el tipo (Movie → "Película", TV Show → "Serie").

dictionary = {
    'Movie'  : 'Pelicula',
    'TV Show': 'Serie' 
}

df_disney['type'] = df_disney['type'].map(dictionary)
df_disney
```

```python
# 7.Eliminar duplicados basados en title.

df_disney = df_disney.drop_duplicates(subset=['title'])
df_disney
```

```python
# 8. Agrupar por release_year y contar cuántos títulos se lanzaron por año.

# Realizo una copia del Data Frame "df_disney"
df_disney2 = df_disney.copy()
count_titleForYears = df_disney2.groupby(['release_year'])['title'].count()
print("La cantidad de Peliculas que se lanzaron por años son:\n",count_titleForYears)
```

```python
# 9. Crear una tabla pivote con type como filas y rating como columnas, mostrando cantidad de títulos.
count_title = df_disney['title'].count()
df_disneyPivot = df_disney.pivot_table(values = 'title', index ='type', columns ='rating', aggfunc ='count', fill_value=0)
df_disneyPivot
```

```python
# 10. Ordenar los títulos por año de lanzamiento de más nuevo a más viejo.
df_disney.sort_values(by = 'release_year', ascending = False, inplace = True)
df_disney
```
