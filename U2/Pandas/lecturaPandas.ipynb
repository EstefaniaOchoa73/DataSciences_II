```python
# Import libraries
import pandas as pd
```

```python
# Lectura de archivos planos
# Sin cabecera.
df_txt = pd.read_csv('text.txt',sep = ' ', header = None) 

# Con cabecera
df_txt = pd.read_csv('text.txt',sep = ' ', names = ['c1','c2','c3','c4','message'])
df_txt
```

```python
# Lectura de archivos planos (FECHAS)
df_date = pd.read_csv('textDate.txt',sep = ',')
print('DataFrame sin castear:',df_date.dtypes)
df_date

# Casteamos la colmuna date a tipo fecha
df_date = pd.read_csv('textDate.txt',sep = ',', parse_dates = ['date'])
print('DataFrame casteado:',df_date.dtypes)
df_date
```

```python
# Lectura de archivos planos (BOOLEANOS)
df_bool = pd.read_csv('textBool.txt',sep = ',')
print('Tipo de Dato del DataFrme:',df_bool.dtypes)
df_bool = pd.read_csv('textBool.txt',sep = ',',dtype = {'asistio':bool,
                                                        'tarea':bool})
print('Tipo de Dato del DataFrme:',df_bool.dtypes)
```
