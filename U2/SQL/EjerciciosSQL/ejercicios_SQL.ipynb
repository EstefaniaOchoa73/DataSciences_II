```python
# 1- Mostrar todos los clientes.

df_Customers = pd.read_sql_query('SELECT CustomerID, ContactName, Address FROM Customers', conn)
df_Customers
```

```python
# 4-Listar las √≥rdenes realizadas por clientes de Alemania.

df_Orders = pd.read_sql_query('SELECT Orders.OrderID, Orders.ShipName FROM Orders ' \
                              'INNER JOIN Customers ON Orders.CustomerID = Customers.CustomerID ' \
                              'WHERE Customers.Country = "Germany"',conn)
df_Orders
```

```python
# 5-Obtener todos los productos con precio mayor a 20.
df_products20 = pd.read_sql_query('SELECT Products.* FROM Products WHERE Products.UnitPrice > 20',conn)
df_products20
```

```python
# 6-Mostrar los nombres y cargos de los empleados.

df_employees = pd.read_sql_query('SELECT Employees.LastName, Employees.FirstName, Employees.Title ' \
                                 'FROM Employees',conn)
df_employees
```

```python
# 7-Listar los nombres de los clientes y su pa√≠s.

df_clients = pd.read_sql_query('SELECT ContactName, Country FROM Customers',conn)
df_clients
```

```python
# 8-Contar cu√°ntos productos hay en total.

df_productsCount = pd.read_sql_query('SELECT COUNT(ProductID) FROM Products',conn)
df_productsCount
```

```python
# 9-Mostrar los productos ordenados por precio de mayor a menor.

df_productsOrder = pd.read_sql_query('SELECT ProductName, UnitPrice FROM Products ORDER BY UnitPrice DESC',conn)
df_productsOrder
```

```python
# 10-Mostrar las categor√≠as de productos.

df_category = pd.read_sql_query('SELECT * FROM Categories',conn)
df_category 
```

```python
# 11-Listar los productos y el nombre de su categor√≠a.
# Dos formas posibles.

df_prodCat = pd.read_sql_query('SELECT Products.ProductName, Categories.CategoryName ' \
                              'FROM Products ' \
                               'INNER JOIN  Categories ' \
                               'ON Products.CategoryID == Categories.CategoryID',conn)

df_prodCat = pd.read_sql_query('SELECT Products.ProductName, Categories.CategoryName ' \
                               'FROM Products, Categories ' \
                               'WHERE Products.CategoryID = Categories.CategoryID',conn)
df_prodCat
```

```python
# 12-Mostrar los productos cuyo nombre comience con la letra 'C'.

df_productC = pd.read_sql_query('SELECT Products.*' \
                                'FROM Products WHERE ProductName LIKE "C%"',conn)

df_productC
```

```python
# 13-Ver los proveedores de productos.

df_vendors = pd.read_sql_query('SELECT Products.ProductName, Suppliers.CompanyName, Suppliers.ContactName, Suppliers.ContactTitle, Suppliers.Address ' \
                               'FROM Suppliers ' \
                               'INNER JOIN Products ' \
                               'ON Suppliers.SupplierID = Products.SupplierID',conn)
df_vendors
```

```python
# 14-Ver todos los pedidos hechos en el a√±o 1997.

df_orders = pd.read_sql_query('SELECT OrderID FROM Orders WHERE OrderDate LIKE "1997%"',conn)
#df_orders = pd.read_sql_query('SELECT * FROM Orders',conn)
df_orders
```

```python
# 15-Mostrar los pa√≠ses √∫nicos de los clientes.

# SELECT DISTINCT retorna los valores distintos o diferentes.

df_countryClients = pd.read_sql_query('SELECT DISTINCT Country FROM Customers',conn)
df_countryClients
```

```python
# 16-Ver la cantidad de productos por categor√≠a.
df_categoryProd = pd.read_sql_query('SELECT ProductName, COUNT(ProductID)' \
                                    'FROM Products ' \
                                    'INNER JOIN Categories ON Products.CategoryID = Categories.CategoryID',conn)
df_categoryProd
```

```python
# 17-Mostrar las √≥rdenes junto con el nombre del empleado que las gestion√≥.
df_orders = pd.read_sql_query('SELECT Orders.OrderID, Employees.FirstName, Employees.LastName ' \
                              'FROM Orders ' \
                              'INNER JOIN Employees ' \
                              'ON Orders.EmployeeID = Employees.EmployeeID ',conn)
df_orders
```

```python
# 18-Ver qu√© productos tienen stock

df_stock = pd.read_sql_query('SELECT Products.* FROM Products WHERE UnitsInStock != 0',conn)
df_stock
```

```python
# 19-Calcular el precio promedio de los productos.

df_priceAvg = pd.read_sql_query('SELECT ProductName AS ProductName, AVG(QuantityPerUnit) AS PrecioPromedio FROM Products',conn)
df_priceAvg
```

```python
# 20-Mostrar los 5 productos m√°s caros.

df_priceMax = pd.read_sql_query('SELECT TOP ProductName, UnitPrice FROM Products ORDER BY UnitPrice',conn)
df_priceMax
```
