```python
# DISNEY PRACTICES

# Importar librerÃ­as
import pandas as pd

# Leer el .csv
df_disney = pd.read_csv('Copia-de-disney_plus_titles.csv')
df_disney
```

```python
# Obtener informaciÃ³n del DataSet.
print("--DESCRIPCIÃ“N DEL DATASET--")
print(df_disney.describe())
print("--INFO DEL DATASET--")
print(df_disney.info())
```

```python
# 1.Mostrar cuÃ¡ntos valores nulos hay por columna.
print("Â¿Cuantos Valores Nulos hay?")
print(df_disney.isna().sum())
```

```python
# 2.Reemplazar valores nulos en rating por "No Rating".
df_disney['rating'] = df_disney['rating'].fillna('No Rating')
```

```python
# 3.Filtrar todos los tÃ­tulos agregados a la plataforma en el aÃ±o 2021.
titles_2021 = df_disney.loc[df_disney['release_year'] == 2021]
titles_2021
```

```python
# 4.Contar cuÃ¡ntos tÃ­tulos hay de cada tipo (Movie, TV Show).

# El atributo Shape permite conocer la dimensiÃ³n de un DataFrame. 
# Shape[0] = Cantidad de filas.
# Shape[1] = Cantidad de columnas.
count_Movie = df_disney.loc[df_disney['type'] == 'Movie'].shape[0]
print("La cantidad de tÃ­tulos del tipo Movie son:", count_Movie)
```

```python
# 5.Crear una nueva columna que indique si el tÃ­tulo fue lanzado antes o despuÃ©s del aÃ±o 2000.
df_disney['Lanzamiento'] = df_disney['release_year'].apply(lambda x: 'After' if x > 2000 else 'Before')
df_disney
```

```python
# 6.Usar .map() para traducir el tipo (Movie â†’ "PelÃ­cula", TV Show â†’ "Serie").

dictionary = {
    'Movie'  : 'Pelicula',
    'TV Show': 'Serie' 
}

df_disney['type'] = df_disney['type'].map(dictionary)
df_disney
```

```python
# 7.Eliminar duplicados basados en title.

df_disney = df_disney.drop_duplicates(subset=['title'])
df_disney
```

```python
# 8. Agrupar por release_year y contar cuÃ¡ntos tÃ­tulos se lanzaron por aÃ±o.

# Realizo una copia del Data Frame "df_disney"
df_disney2 = df_disney.copy()
count_titleForYears = df_disney2.groupby(['release_year'])['title'].count()
print("La cantidad de Peliculas que se lanzaron por aÃ±os son:\n",count_titleForYears)
```

```python
# 9. Crear una tabla pivote con type como filas y rating como columnas, mostrando cantidad de tÃ­tulos.
count_title = df_disney['title'].count()
df_disneyPivot = df_disney.pivot_table(values = 'title', index ='type', columns ='rating', aggfunc ='count', fill_value=0)
df_disneyPivot
```

```python
# 10. Ordenar los tÃ­tulos por aÃ±o de lanzamiento de mÃ¡s nuevo a mÃ¡s viejo.
df_disney.sort_values(by = 'release_year', ascending = False, inplace = True)
df_disney
```
